```python
#generate top 20 highest cost by products
summary = df_all_transactions.groupby(['DrugName']).agg({'Final_Cost': 'sum'}).sort_values(by='Final_Cost', ascending=False).head(20)

print(summary)
```

```python
# Histogram for Non-Injections Return Qty 

plt.hist(df_all_transactions[df_all_transactions.SubCat != 'INJECTIONS'].ReturnQuantity, edgecolor = 'green', color = 'yellow')
plt.xlabel('Returned Qty per Transaction')
plt.ylabel('Total Returned Quantity')
plt.title('Histogram  Non-Injections Return Quantity  Jan-Dec 2022')
plt.show()
```

```python
# Histogram for Non-Injections Sales 

plt.hist(df_all_transactions[df_all_transactions.SubCat != 'INJECTIONS'].Final_Sales, edgecolor = 'green', color = 'yellow')
plt.xlabel('Sales per Transaction')
plt.ylabel('Total Final Sales')
plt.title('Histogram  Non-Injections Final Sales  Jan-Dec 2022')
plt.show()
```

```python
# Histogram for Non-Injections RtnMrp 

plt.hist(df_all_transactions[df_all_transactions.SubCat != 'INJECTIONS'].RtnMRP, edgecolor = 'green', color = 'yellow', range=(0,4500))
plt.xlabel('RtnMRP per Transaction')
plt.ylabel('Total RtnMRP')
plt.title('Histogram  Non-Injections RtnMRP  Jan-Dec 2022')
plt.show()
```
